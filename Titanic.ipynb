```python
import pandas as pd   #importing pandas

df = pd.read_csv("titanic.csv")
df
```

```python
print(df.shape)   # Displaying rows and columns
print(df.info())   # Information in dataframe
df.describe()   # Statistics of dataset
```

```python
print(df.isna().sum(), "\n")   # checking the missing values
```

```python
from sklearn import preprocessing   #importing preprocessing

label_encoder = preprocessing.LabelEncoder()   #to convert sex data to numeric
df[["sex"]] = label_encoder.fit_transform(df["sex"])
df["sex"].value_counts()
```

```python
df  = df.drop(["name", "cabin", "fare", "home.dest"], axis = 1)   # Droping unnecessory data from dataframe
df
```

```python
df["age"] = df["age"].fillna(df["age"].mean())   #replacing missing values with mean
df
```

```python
df["boat"] = df["boat"].fillna(df["boat"].mode()[0])   #replacing missing values with mode
df
```

```python
df["body"] = df["body"].fillna(df["body"].median())   #replacing missing values with median
df
```

```python
g = df.groupby("survived")   #groping the survived data
g["embarked"].value_counts()   #counting the embarked based on survived
```

```python
df["embarked"] = df["embarked"].fillna(value = "s")   #replacing missing values with s
df
```

```python
df["embarked"] = label_encoder.fit_transform(df["embarked"])   #to convert embarked data to numeric
df["embarked"].value_counts()
```

```python
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)

from matplotlib import pyplot as plt
import seaborn as sns
sns.countplot(df["embarked"], hue = df["survived"])
plt.show
```

```python
df.corr()   #pairwise correlation of all columns in the dataframe
```

```python
df.plot(x = "embarked", y = ["sibsp", "parch"], kind = "bar")   #ploting graph between sibsp and parch
plt.title("bar graph")
```

```python
sns.heatmap(df.corr())   #representing various shades of the same colour for each value to be plotted
```

```python
df["family"] = df["sibsp"] + df["parch"] + 1   #grouping sibsp and parch as a sigle column
df = df.drop(["sibsp", "parch", "ticket", "boat"], axis = 1)   #droping the unwanted data
df
```

```python
from sklearn.model_selection import train_test_split

# Segregating labels into separate variable
x = df.drop("survived", axis = 1).values
y = df["survived"].values

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=6)
```

```python
from sklearn.preprocessing import StandardScaler   #importing StandardScaler

sc = StandardScaler()   # Assigning StandardScaler to sc
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression   # import LogisticRegression

log_reg = LogisticRegression()   # Assigning LogisticRegression() to log_reg
log_reg.fit(x_train, y_train)   #fitting log_reg to the train the model
```

```python
from sklearn.metrics import accuracy_score   # import accuracy_score

predict = log_reg.predict(x_test)   #predict instances from the test set

# finding score for logreg model
print("accuracy of logistic regression classifert: ", log_reg.score(x_test, y_test))

accuracy_score(y_test, predict)
```

```python
from sklearn.metrics import confusion_matrix # import confusion_matrix

predict = log_reg.predict(x_test)   #predict instances from the test set

confusion_matrix(y_test, predict)
```
