```python
import pandas as pd   #importing pandas

df = pd.read_csv("creditcard.csv")   #reading CSV file
df.head()   #displaying first 5 rows
```

```python
print(df.shape)   # Displaying rows and columns
print(df.info())   # Information in dataframe
df.describe()   # Statistics of dataset
```

```python
print(df.isna().any())   # checking missing values

# Filling missing values with most freaquently occured values
df["Approvalstatus"] = df["Approvalstatus"].fillna(df["Approvalstatus"].mode()[0])
df.isnull()   # Checking the missing values
```

# Droping unnecessory data from dataframe

```python
df = df.drop(["married", "bankCustmer", "eductionalLevel", "enthnicity", "DriverLicense", "citizen"], axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder   #import LabelEncode
label = LabelEncoder()   # instantiate LabelEncoder

for col in df.columns.values:   #iterate over each column of cc_apps
    if df[col].dtypes == "object":   #check if the column is of object type
        df[col] = label.fit_transform(df[col])   #use LabelEncoder to do the numeric transformation
        
df[col]
```

# Percentage of Fraud & Non-Fraud transactions

```python
fraud = 0
non_fraud = 0
status = df["Approvalstatus"]
for i in range(len(status)):
    if status[i]==1:
        fraud+=1
    else:
        non_fraud+=1
        
print("Total num of non fraud transactions is: ", non_fraud)
non_fraud_per = (non_fraud/len(status))*100
print("percentage of non fraud transactions in the data set is: ", non_fraud_per, "\n")

print("Total num of non fraud transactions is: ", fraud)
fraud_per = (fraud/len(status))*100
print("percentage of non fraud transactions in the data set is: ", fraud_per)
```

# For modeling we need to split dataset
    0.8 = 80% of data will use to train the model
    0.2 = 20% of data will validate the model

```python
from sklearn.model_selection import train_test_split

# Segregating labels into separate variable
x = df.drop(["Approvalstatus"], axis=1)
y = df["Approvalstatus"]

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler   #importing StandardScaler

sc = StandardScaler()   # Assigning StandardScaler to sc
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression   # import LogisticRegression

log_reg = LogisticRegression()   # Assigning LogisticRegression() to log_reg
log_reg.fit(x_train, y_train)   #fitting log_reg to the train the model
```

```python
from sklearn.metrics import confusion_matrix # import confusion_matrix

y_predict = log_reg.predict(x_test)   #predict instances from the test set

# finding score for logreg model
print("accuracy of logistic regression classifert: ", log_reg.score(x_test, y_test))

confusion_matrix(y_test, y_predict)
```
