```python
import pandas as pd   # Importing pandas

movie = pd.read_csv("movies.csv")   # Reading CSV file
movie.head()   # Displaying first 5 rows
```

```python
for i in movie.columns:   # printing all columns
    print(i)

movie.shape  # Displaying rows and columns
```

```python
movie.describe()   # Statistic of the dataframe
```

```python
movie["original_language"].value_counts()   # Count of languages
```

```python
import matplotlib.pyplot as plt   # Importing matplotlib

plt.bar(list(movie["original_language"].value_counts()[0:5].keys()), list(movie["original_language"].value_counts()[0:5]), color=["red", "blue", "orange", "green", "yellow"])
plt.title("Orginal Languages")
plt.xlabel("Languages")
plt.ylabel("Num of Movies")
plt.show()   # Graph for num of movies in first 5 languages
```

```python
plt.hist(movie["vote_average"], color=["m"])   #voting for the movies
plt.title("Voting")
plt.xlabel("Rating")
plt.ylabel("Num of movies")
plt.show()
```

```python
high_rated_movies = movie[movie["vote_average"]>8]   # Movies have rating more than 8
print(high_rated_movies.shape)   # Rows and columns have rating more than 8
high_rated_movies.head()   # First 5 rows have rating more than 8
```

```python
top_5 = movie.sort_values(by="vote_average", ascending=False).head()
top_5   #top 5 movies have top rating
```

```python
top_5_revenue = movie.sort_values("revenue", ascending=False).head()
top_5_revenue   #top 5 movies have top revenue
```

```python
plt.hist(movie["runtime"], color="c")
plt.show()   # To check how long the num of movies were in threaters
```
