    The documation tells us the feature 'CIRM' is per captia crime rate
    'NX' is nitric oxide concentration and 'RM' average number of room per dwelling
    eg: The target variable 'MEDV', is the median value of owner occupied homes in thousand of dollars

```python
import pandas as pd   #importing pandas

boston = pd.read_csv("boston.csv")    #reading dataset
boston.head()    #displaying first 5 rows
```

```python
print(boston.shape)   # Displaying rows and columns
print(boston.info())   # Information in dataframe
boston.describe()   # Statistics of dataset
```

```python
x = boston.drop("MEDV", axis = 1).values   #creating feature array
y = boston["MEDV"].values      #creating traget array
print(x)
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression   #importing LinearRegression

reg = LinearRegression()   #Instantiate the regression
reg.fit(x_train, y_train)    #fitting the traning data

y_pred = reg.predict(x_test)   #predicting y
reg.score(x_test, y_test)    #default scoring method for linear regression
```

```python
import matplotlib.pyplot as plt    #importing matplotlib
import numpy as np

plt.scatter(y_test, y_pred, color = "blue")   #drawing scatter plot
z = np.polyfit(y_test, y_pred, 1)
p = np.poly1d(z)

plt.plot(y_test, p(y_test), "g--")
plt.show()

#distance b/w point and line is the error it is also known as reseduals
```

```python
from sklearn.model_selection import cross_val_score   #importing cross valdation
import numpy as np   #importing numpy

cv_results = cross_val_score(reg, x, y, cv=5)   #cross valdation for 1st 5
print(cv_results)

np.mean(cv_results)    #mean of the cross valdation
```
